# Problema de otimização

## Formulação quadrática

$$\begin{aligned}
\min_{x \in \{0, 1\}^n}  x^T \Sigma x - \mu^T x\\
\text{subject to: } A x = L \\
                    R^T x \leq c
\end{aligned}$$



- $x \in \{0, 1\}^n$ vetor binário que indica se o ativo foi escolhido ($x[i] = 1$) ou não ($x[i] = 0$),
- $\mu \in \mathbb{R}^n$ retorno médio dos ativos (`lista_produto.ret_med`),
- $\Sigma \in \mathbb{R}^{n \times n}$ matriz de coravariância (`matriz_de_covariancia`),
- $A$ Matriz do vínculo que define quantidade de produtos por classe,
- $L$ vetor com quantidade de produtos que devem ser selecionados de cada classe (`dict_quantidade_produtos`).
- $R$ matriz que representa a avaliação de risco dos produtos (`lista_produto.aval`).
- $c$ máximo risco permitido

# Cenário clássico

## Configurações

### Observação: Favor não alterar nada na célula de código abaixo!

````python
lista_produtos = pd.read_csv(filepath_or_buffer="dados_batalha_classico.csv", index_col=0)
matriz_de_covariancia = pd.read_csv(filepath_or_buffer="covariancia_classico.csv", index_col=0)

dict_quantidade_produtos = {'classe_0': 5,
                            'classe_1': 2,
                            'classe_2': 3,
                            'classe_3': 1,
                            'classe_4': 1,
                            'classe_5': 4,
                            'classe_6': 2,
                            'classe_7': 2,
                            'classe_8': 1,
                            'classe_9': 1}

c = 50
````

#### Solução Clássica: CPLEX, XPRESS

````python
# desenvolva seu código abaixo

# Resolvemos o problema clássico usando a biblioteca cvxpy
import cvxpy as cp
import numpy as np
import pandas as pd

num_prod = len(lista_produtos)
num_classes = len(dict_quantidade_produtos.values())

# Definimos a variável do problema
x = cp.Variable(shape=num_prod, boolean=True)

A = np.zeros((num_classes,num_prod))

# lista com as classes
classes = list(lista_produtos['classe'])

# cria a matrix "A" da restrição
for i,cl in enumerate(classes):
    line = int(cl[::-1][0])
    A[line, i] = 1

# cria o vetor "L" da restrição
L = list(dict_quantidade_produtos.values())

# cria o vetor "R" da restrição
R = np.array(list(lista_produtos['aval'])).T

#Define a função a ser minimizada
objective = cp.Minimize(x.T@matriz_de_covariancia@x - np.array(lista_produtos["ret_med"]).T@x)

#Define as restrições do problema
constraints = [A@x == L]
constraints += [R@x <= c]

#Resolve o problema usando CPLEX e XPRESS
problem = cp.Problem(objective=objective, constraints=constraints)
solvers = ["CPLEX", "XPRESS"]
resultados = []
tempos_execucao = []
produtos = []

P = list(lista_produtos['produto'])

for solver in solvers:
    resultado = problem.solve(solver=solver, verbose=False)
    resultados.append(resultado)
    tempos_execucao.append(problem.solver_stats.solve_time)
    
    solutions = list(problem.solution.primal_vars.values())[0]
    tmp_list = []
    lst = []
    for s in range(len(solutions)):
        if int(solutions[s]) == 1:
            lst.append(s)
            
    for p in range(len(lst)):
        tmp_list.append(P[lst[p]])
    
    produtos.append(tmp_list)  
    
df = pd.DataFrame()
df["Solver"] = solvers
df["Tempo de Execucao"] = tempos_execucao
df["Valor ótimo"] = resultados
df["Produtos"] = produtos
df
````

#### Solução Quântica: Quantum Approximate Optimization Algorithm (QAOA)

````python
# Faz circuito do hamiltoniano de custo
def cost_hamiltonian_circuit(quantumCircuit, num_vars, linear, quadratic, gamma):
    #Termos quadráticos do Hamiltoniano de custo
    ct = 0
    for i in range(num_vars):
        for j in range(num_vars):
             if(i<j):
                # CNOT, RZ, CNOT
                quantumCircuit.cx(i,j)
                quantumCircuit.rz(quadratic[ct]*gamma,j)
                quantumCircuit.cx(i,j)
                quantumCircuit.barrier()
                ct+=1
    quantumCircuit.barrier()
    #Termos lineares do Hamiltoniano de custo
    ct = 0
    for i in range(num_vars):
        # RZ
        quantumCircuit.rz(linear[ct]*gamma,i)
        ct+=1
    quantumCircuit.barrier()

    return quantumCircuit

# Faz circuito do Hamiltoniano mixer
def mixer_hamiltonian_circuit(quantumCircuit, num_vars,beta):
    for i in range(num_vars):
        quantumCircuit.ry(beta, i)
    for i in range(num_vars-1):
        quantumCircuit.cx(i, i+1)
    quantumCircuit.cx(num_vars-1, 0)
    for i in range(num_vars):
        quantumCircuit.ry(beta, i)
    quantumCircuit.barrier() 
    return quantumCircuit

# Função que retorna o circuito qaoa completo
def qaoa_circuit(num_vars, linear, quadratic, layer_number):
    # Cria circuito
    qaoa_circ = QuantumCircuit(num_vars)
    gammas = ParameterVector(name=r"$\gamma$", length=layer_number)
    betas = ParameterVector(name=r"$\beta$", length=layer_number)
    
    #estado inicial do circuito
    for i in range(num_vars):
        qaoa_circ.h(i)
    qaoa_circ.barrier()

    #roda o loop do qaoa
    for l in range(layer_number):
        qaoa_circ = cost_hamiltonian_circuit(qaoa_circ, num_vars, linear,quadratic, gammas[l])
        qaoa_circ = mixer_hamiltonian_circuit(qaoa_circ, num_vars, betas[l])
    
    return qaoa_circ

# Função que calcula do valor esperado do hamiltoniano de custo (função de custo)
def cost_function_qaoa_120(params):
    
    qaoa_circ= qaoa_circuit(num_vars=num_vars_120, linear=linear_120, quadratic=quadratic_120, layer_number=layer_number_120)
    qaoa_circ = qaoa_circ.assign_parameters(params)
    backend = Aer.get_backend("statevector_simulator")
    resultado = backend.run(qaoa_circ).result()
    opt_state = resultado.get_statevector()
    #calcula valor esperado do hamiltoniano de custo usando o estado medido no circuito qaoa
    cost_function_value = np.real(np.array(opt_state)@op_120.to_matrix()@np.array(opt_state))

    return cost_function_value
````

````python
# parametros para criar o hamiltoniano desse seed
seed_71 = seed_list[0]
sample, cov_sample = prepara_amostra_problema(seed=seed_71)
qubo_71 = carrega_problema_qubo(seed=seed_71)
num_vars_71 = qubo_71.get_num_vars()

linear_71, quadratic_71, offset_71 = qubo_to_ising(qubo_71)

op_71 = get_cost_hamiltonian(num_vars_71, linear_71, quadratic_71, offset_71)

print(op_71)
````

````python
#define função de custo que será minimizada pelo qaoa para esse seed
def cost_function_qaoa_71(params):
    
    qaoa_circ= qaoa_circuit(num_vars=num_vars_71, linear=linear_71, quadratic=quadratic_71, layer_number=layer_number_71)
    qaoa_circ = qaoa_circ.assign_parameters(params)
    backend = Aer.get_backend("statevector_simulator")
    resultado = backend.run(qaoa_circ).result()
    opt_state = resultado.get_statevector()
    
    cost_function_value = np.real(np.array(opt_state)@op_71.to_matrix()@np.array(opt_state))

    return cost_function_value
````
